# Question 1: Load a CSV Dataset into a Pandas DataFrame

In this step, the CSV dataset is loaded into a Pandas DataFrame using Pandas library.

```python
# Loading CSV dataset

df = pd.read_csv("data.csv")

print("Dataset loaded successfully!")
```

# Question 3: Handle Missing Values (identify, fill/drop)

In this step, missing values are identified and handled using appropriate methods.

```python
# Checking missing values in dataset

print(df.isnull().sum())
```

```python
# Selecting numeric columns

numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns

print("Numeric Columns:")
print(numeric_columns)
```

```python
# Filling numeric missing values with mean

for col in numeric_columns:
    
    df[col] = df[col].fillna(df[col].mean())

print("Numeric missing values handled successfully!")
```

```python
# Selecting categorical columns

categorical_columns = df.select_dtypes(include=['object']).columns

print("Categorical Columns:")
print(categorical_columns)
```

```python
# Filling categorical missing values with mode

for col in categorical_columns:
    
    df[col] = df[col].fillna(df[col].mode()[0])

print("Categorical missing values handled successfully!")
```

```python
# Checking missing values again after handling

print(df.isnull().sum())
```

```python
# Handling remaining missing values

df = df.fillna("Not Available")

print(df.isnull().sum())
```

# Question 4: Perform Basic Operations (filter rows, select columns)

In this step, basic operations such as selecting columns and filtering rows are performed on the dataset.

```python
# Selecting first two columns

selected_columns = df.iloc[:, :2]

print(selected_columns.head())
```

```python

# Filtering rows using numeric column

numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

filtered_data = df[df[numeric_cols[0]] > df[numeric_cols[0]].mean()]

print(filtered_data.head())
```

# Question 5: Remove Duplicates

In this step, duplicate rows are identified and removed from the dataset.

```python
# Checking duplicate rows

print("Duplicate Rows:", df.duplicated().sum())
```

```python
# Removing duplicate rows

df = df.drop_duplicates()

print("Duplicates removed successfully!")
```

# Question 6: Create a Derived Column (total_amount = price * quantity)

In this step, a new column named total_amount is created using price and quantity columns.

```python
# Creating temporary price and quantity columns

df['price'] = df['initial_price']

df['quantity'] = df['ratings_count']
```

```python
# Creating derived column

df['total_amount'] = df['price'] * df['quantity']

print("Derived column created successfully!")

print(df[['price', 'quantity', 'total_amount']].head())
```

# Question 7: Save the Cleaned Dataset as a New CSV File

In this step, the cleaned dataset is saved as a new CSV file for future analysis.

```python
# Saving cleaned dataset

df.to_csv("cleaned_data.csv", index=False)

print("Cleaned dataset saved successfully!")
```
